```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import xgboost as xgb
import seaborn as sns
#from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import StratifiedKFold
from sklearn.preprocessing import Normalizer
%matplotlib inline
```

```python
data = pd.read_csv("train.csv")
data_test = pd.read_csv("test.csv")
```

## Correlation Matrix

Eliminating data without output-correlation

```python
unwanted_colums = data.columns[data.columns.str.startswith('ps_calc_')]
data = data.drop(unwanted_colums, axis = 1)
data_test = data_test.drop(unwanted_colums, axis = 1)
```

```python
# Compute the correlation matrix
corr = data.corr()

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(220, 10, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
plt.show()
```

Generate the dummy variables for categorical data

```python
bin_cols = [col for col in data if col.endswith('bin')]
cat_cols = [col for col in data if col.endswith('cat')]

data = pd.get_dummies(data, columns=cat_cols, drop_first=True)
data_test = pd.get_dummies(data_test, columns=cat_cols, drop_first=True)
```

Stratified Kfold

```python
kfold = 5
skf = StratifiedKFold(n_splits=kfold, random_state=42)
```

XGBoost params

```python
params = {
    'min_child_weight': 10.0,
    'objective': 'binary:logistic',
    'max_depth': 7,
    'max_delta_step': 1.8,
    'colsample_bytree': 0.4,
    'subsample': 0.8,
    'eta': 0.025,
    'gamma': 0.65,
    'num_boost_round' : 700
    }
```

Gini Metrics

```python
def gini(actual, pred, cmpcol = 0, sortcol = 1):
    assert( len(actual) == len(pred) )
    all = np.asarray(np.c_[ actual, pred, np.arange(len(actual)) ], dtype=np.float)
    all = all[ np.lexsort((all[:,2], -1*all[:,1])) ]
    totalLosses = all[:,0].sum()
    giniSum = all[:,0].cumsum().sum() / totalLosses
    
    giniSum -= (len(actual) + 1) / 2.
    return giniSum / len(actual)
 
def gini_normalized(a, p):
    return gini(a, p) / gini(a, a)

def gini_xgb(preds, dtrain):
    labels = dtrain.get_label()
    gini_score = gini_normalized(labels, preds)
    return 'gini', gini_score
```

```python
X = data.drop(['id', 'target'], axis=1).values
y = data.target.values
test_id = data_test.id.values
test = data_test.drop('id', axis=1)
```

Normalizing data

```python
norm = Normalizer()
norm.fit(X)
```

```python
X = norm.transform(X)
test_val = norm.transform(test.values)
```

training

```python
sub = pd.DataFrame()
sub['id'] = test_id
sub['target'] = np.zeros_like(test_id)
del data
del data_test
```

```python
for i, (train_index, test_index) in enumerate(skf.split(X, y)):
    print(test_index)
```

```python
for i, (train_index, test_index) in enumerate(skf.split(X, y)):
    print('[Fold %d/%d]' % (i + 1, kfold))
    X_train, X_valid = X[train_index], X[test_index]
    y_train, y_valid = y[train_index], y[test_index]
    # Convert our data into XGBoost format
    d_train = xgb.DMatrix(X_train, y_train)
    d_valid = xgb.DMatrix(X_valid, y_valid)
    d_test = xgb.DMatrix(test_val)
    watchlist = [(d_train, 'train'), (d_valid, 'valid')]

    # Train the model! We pass in a max of 1,600 rounds (with early stopping after 70)
    # and the custom metric (maximize=True tells xgb that higher metric is better)
    mdl = xgb.train(params, d_train, 1600, watchlist, early_stopping_rounds=70, feval=gini_xgb, maximize=True, verbose_eval=100)

    print('[Fold %d/%d Prediciton:]' % (i + 1, kfold))
    # Predict on our test data
    p_test = mdl.predict(d_test, ntree_limit=mdl.best_ntree_limit)
    sub['target'] += p_test/kfold
```

```python
sub.to_csv('StratifiedKFold.csv', index=False)
```
